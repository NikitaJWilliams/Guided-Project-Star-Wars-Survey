# Guided Project: Star Wars Survey

## Introduction

In this project, I will be analyzing the *Star Wars* movies survey data. Prior to the release of *Star Wars: The Force Awakens*, the team at FiveThirtyEight wondered what is the most popular and highest rate *Star Wars* film. To answer the question, they surveyed *Star Wars* fans with the help of the online tool SurveyMonkey. They gathered 835 responses which can be accessed through this [link](https://github.com/fivethirtyeight/data/tree/master/star-wars-survey).

## Preliminary Data Analysis

```python
import pandas as pd
import numpy as np 

star_wars = pd.read_csv("star_wars.csv", encoding = "ISO-8859-1")

star_wars.head(10)
```

```python
star_wars.columns
```

## Data Cleaning

**RespondentID** column stands for a unique ID for each respondent; however, some values in the column are missing. I will remove any rows with any blank cells in the **RespondentID** column.

```python
star_wars = star_wars[pd.notnull(star_wars['RespondentID'])]

star_wars.head(10)
```

Let's take a look at the next two columns in our dataframe, which are:
- 'Have you seen any of the 6 films in the Star Wars franchise?'
- 'Do you consider yourself to be a fan of the Star Wars film franchise'

Both columns comprises of three unique values: "Yes", "No" and NaN where NaN stands for a missing value. To facilitate my data analysis down the road, I will transform the two columns into Boolean type. 

```python
booln = {"Yes":True,"No":False}

star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(booln)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(booln) 

star_wars.head()
```

The next six columns serve as a single checkbox question: "Which of the following Star Wars films have you seen? Please select all that apply". These columns contain two possible values: the name of the movie in question and NaN. If the value in a cell is the name of the movie in question, then that means that the respondent has seen that film. I will assume that NaN represents a negative response to the question.

Just like before, I will cnovert each of the six columns to a Boolean type and then rename the columns to something more intuitive.

```python
s1 = {"Star Wars: Episode I  The Phantom Menace": True, np.NaN:False}
s2 = {"Star Wars: Episode II  Attack of the Clones": True, np.NaN:False}
s3 = {"Star Wars: Episode III  Revenge of the Sith": True, np.NaN:False}
s4 = {"Star Wars: Episode IV  A New Hope": True, np.NaN:False}
s5 = {"Star Wars: Episode V The Empire Strikes Back": True, np.NaN:False}
s6 = {"Star Wars: Episode VI Return of the Jedi": True, np.NaN:False}

star_wars["Which of the following Star Wars films have you seen? Please select all that apply."] = star_wars["Which of the following Star Wars films have you seen? Please select all that apply."].map(s1)
star_wars['Unnamed: 4'] = star_wars['Unnamed: 4'].map(s2)
star_wars['Unnamed: 5'] = star_wars['Unnamed: 5'].map(s3)
star_wars['Unnamed: 6'] = star_wars['Unnamed: 6'].map(s4)
star_wars['Unnamed: 7'] = star_wars['Unnamed: 7'].map(s5)
star_wars['Unnamed: 8'] = star_wars['Unnamed: 8'].map(s6)

star_wars = star_wars.rename(columns = {star_wars.columns[3]:"seen_1",star_wars.columns[4]:"seen_2",star_wars.columns[5]:"seen_3",star_wars.columns[6]:"seen_4",star_wars.columns[7]:"seen_5",star_wars.columns[8]:"seen_6"})

star_wars.head()
```

The next six columnns ask the respondent to rank the *Star Wars* from most favorite to least favorite in the series on a scale of 1 to 6. 1 stands for the most favorite film and 6 stands for the least favorite film. In order to perform quantitative analysis, I will convert these columns to numeric type. In addition, I will rename them to something more suitable.

```python
star_wars[star_wars.columns[9:15]]= star_wars[star_wars.columns[9:15]].astype(float)

star_wars = star_wars.rename(columns = {star_wars.columns[9]:"ranking_1",star_wars.columns[10]:"ranking_2",star_wars.columns[11]:"ranking_3",star_wars.columns[12]:"ranking_4",star_wars.columns[13]:"ranking_5",star_wars.columns[14]:"ranking_6"})

star_wars.head()
```

## Data Analysis

Now that I have cleaned the dataset, I am ready to identify the highest-ranked *Star Wars* movie using a bar chart. 

```python
%matplotlib inline

import matplotlib.pyplot as plt

rankings = ["r_1","r_2","r_3","r_4","r_5","r_6"]
y_pos = np.arange(len(rankings))
avg_rankings = [star_wars["ranking_1"].mean(),star_wars["ranking_2"].mean(),star_wars["ranking_3"].mean(),star_wars["ranking_4"].mean(),star_wars["ranking_5"].mean(),star_wars["ranking_6"].mean()]

plt.bar(y_pos,avg_rankings, align = 'center', alpha = 0.5)
plt.xticks(y_pos,rankings)
plt.ylabel("Average Rating")
plt.title("Average Respondent Rating Across 6 Star Wars Movies")  
plt.show()               
```

The bar chart tells us that *Star Wars: Episode V The Empire Strikes Back* is the best film in the franchise so far in the view of the public. An average respondent of the survey clearly prefers the original trilogy over the new films. An overwhelming number of responses rated *Star Wars: Episode III Revenge of the Sith* to be the least favorite movie in the entire series. 

The next step is to determine the most popular *Star Wars* movie in terms of viewership numbers. 

```python
views = ['s_1','s_2','s_3','s_4','s_5','s_6']
v_pos = np.arange(len(rankings))
nviews = [star_wars['seen_1'].sum(),star_wars['seen_2'].sum(),star_wars['seen_3'].sum(),star_wars['seen_4'].sum(),star_wars['seen_5'].sum(),star_wars['seen_6'].sum()]

plt.bar(v_pos, nviews, align = 'center',alpha = 0.5)
plt.xticks(v_pos, views)
plt.ylabel("Number of Views")
plt.title('Number of Respondent Views Across 6 Star Wars Movies')
plt.show()
```

Unsurprisingly, the highest rated film is also the most popular one and the lowest rated film is least popular one in the series. Strikingly, the first movie in the franchise has more views than the fourth one even though its average rating is significantly lower. Perhaps the respondents had high expectations going into the *Star Wars: Episode I The Phantom Menace* following the tremendous success of the previous two movies.

Before I wrap up the project, I want to find out whether male and female respondents have expressed different views on the *Star Wars* movies. 

```python
males = star_wars[star_wars['Gender'] == "Male"]
females = star_wars[star_wars['Gender'] == "Female"]

#highest-ranked movies according to male respondents
mavg_rankings = [males["ranking_1"].mean(),males["ranking_2"].mean(),males["ranking_3"].mean(),males["ranking_4"].mean(),males["ranking_5"].mean(),males["ranking_6"].mean()]

plt.bar(y_pos,mavg_rankings, align = 'center', alpha = 0.5)
plt.xticks(y_pos,rankings)
plt.ylabel("Average Male Rating")
plt.title("Average Male Respondent Rating Across 6 Star Wars Movies")  
plt.show() 
```

```python
#highest-ranked movies according to female respondents
favg_rankings = [females["ranking_1"].mean(),females["ranking_2"].mean(),females["ranking_3"].mean(),females["ranking_4"].mean(),females["ranking_5"].mean(),females["ranking_6"].mean()]

plt.bar(y_pos,favg_rankings, align = 'center', alpha = 0.5)
plt.xticks(y_pos,rankings)
plt.ylabel("Average Female Rating")
plt.title("Average Female Respondent Rating Across 6 Star Wars Movies")  
plt.show()
```

Female and Male respondents agree that the original trilogy is far better than the newer one; however, female respondents dislike the newer trilogy a bit less than the male participants as they seem to not mind *Star Wars: Episode I  The Phantom Menace* and *"Star Wars: Episode II  Attack of the Clones" * too much. 

```python
m_nviews = [males['seen_1'].sum(),males['seen_2'].sum(),males['seen_3'].sum(),males['seen_4'].sum(),males['seen_5'].sum(),males['seen_6'].sum()]

plt.bar(v_pos, m_nviews, align = 'center',alpha = 0.5)
plt.xticks(v_pos, views)
plt.ylabel("Number of Male Views")
plt.title('Number of Male Respondent Views Across 6 Star Wars Movies')
plt.show()
```

```python
f_nviews = [females['seen_1'].sum(),females['seen_2'].sum(),females['seen_3'].sum(),females['seen_4'].sum(),females['seen_5'].sum(),females['seen_6'].sum()]

plt.bar(v_pos, f_nviews, align = 'center',alpha = 0.5)
plt.xticks(v_pos, views)
plt.ylabel("Number of Female Views")
plt.title('Number of Female Respondent Views Across 6 Star Wars Movies')
plt.show()
```

We can infer from the bar chart that the male respondents have watched more *Star Wars* movies than the female ones. The distribution of female views across the six films is far greater than the distribution of male views even though male fans hold lower views of all thew new three films.
